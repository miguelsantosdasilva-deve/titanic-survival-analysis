```python
sobreviventes = df.survived.value_counts()
sobreviventes.index = ['Não sobreviveu', 'Sobreviveu']
sobreviventes.plot(kind="pie", autopct='%.1f%%', colors=['green', 'gray'])
plt.title('Sobreviventes')
plt.show()
```

```python
homens = df[df['sex'] == 'male']
total_homens = len(homens)
contagem_homens = homens.survived.value_counts()
porcentagem_homens = (contagem_homens / total_homens) * 100
porcentagem_homens.index = [f"Falecidos | {porcentagem_homens[0]:.1f}%", f"Sobreviventes | {porcentagem_homens[1]:.1f}%"]
porcentagem_homens.plot(kind='bar', color=['gray', 'green'], rot=0)
plt.grid(axis='y', linestyle=':')
plt.yticks(range(0, 80, 15))
plt.title('Taxa de Sobrevivência - Homens')
plt.show()
```

```python
mulheres = df[df['sex'] == 'female']
total_mulheres = len(mulheres)
contagem_mulheres = mulheres.survived.value_counts()
porcentagem_mulheres = (contagem_mulheres / total_mulheres) * 100
porcentagem_mulheres.index = [f"Sobreviventes | {porcentagem_mulheres[1]:.1f}%", f"Falecidas | {porcentagem_mulheres[0]:.1f}%"]
porcentagem_mulheres.plot(kind='bar', color=['green', 'gray'], rot=0)
plt.grid(axis='y', linestyle=':')
plt.yticks(range(0, 80, 15))
plt.title('Taxa de Sobrevivência - Mulheres')
plt.show()
```

```python
fclass = df[df['pclass'] == 1]
sclass = df[df['pclass'] == 2]
tclass = df[df['pclass'] == 3]
tclass
```

```python
#primeira classe
total_fclass = len(fclass)
contagem_fclass = fclass.survived.value_counts()
porcentagem_fclass = (contagem_fclass / total_fclass) * 100
```

```python
#segunda classe
total_sclass = len(sclass)
contagem_sclass = sclass.survived.value_counts()
porcentagem_sclass = (contagem_sclass / total_sclass) * 100
```

```python
#terceira classe
total_tclass = len(tclass)
contagem_tclass = tclass.survived.value_counts()
porcentagem_tclass = (contagem_tclass / total_tclass) * 100
```

```python
# Versão com cores simples e nomes horizontais
classes_nomes = ['1ª Classe', '2ª Classe', '3ª Classe']
taxas_sobrevivencia = [porcentagem_fclass[1], porcentagem_sclass[1], porcentagem_tclass[1]]

# Usando cores predefinidas
plt.bar(classes_nomes, taxas_sobrevivencia, color=['darkgreen', 'green', 'lightgreen'])

# Comando simples para deixar o texto reto
plt.xticks(rotation=0)

plt.grid(axis='y', linestyle=':')
plt.yticks(range(0, 101, 10))
plt.title('Taxa de Sobrevivência por Classe Social')
plt.ylabel('Porcentagem (%)')
plt.show()
```

```python
sns.boxplot(data=df, x='pclass', y='fare', showfliers=False, palette='Set2')
plt.title('Distribuição de Tarifas por Classe (Sem Outliers)')
plt.xlabel('Classe Social')
plt.ylabel('Preço da Passagem (Fare)')
plt.xticks([0, 1, 2], ['1ª Classe', '2ª Classe', '3ª Classe']) # Nomeia as classes no eixo X
plt.grid(axis='y', linestyle=':')

plt.show()
```

```python
# 1. Separando os grupos por porto
porto_c = df[df['embarked'] == 'C']
porto_s = df[df['embarked'] == 'S']
porto_q = df[df['embarked'] == 'Q']

# 2. Calculando as porcentagens de sobrevivência de cada um (o índice [1] é quem sobreviveu)
taxa_c = (porto_c['survived'].value_counts() / len(porto_c)) * 100
taxa_s = (porto_s['survived'].value_counts() / len(porto_s)) * 100
taxa_q = (porto_q['survived'].value_counts() / len(porto_q)) * 100

# 3. Organizando para o gráfico
portos_nomes = ['Cherbourg (França)', 'Southampton (Inglaterra)', 'Queenstown (Irlanda)']
taxas_finais = [taxa_c[1], taxa_s[1], taxa_q[1]]

# 4. Gerando o gráfico de barras simples
plt.bar(portos_nomes, taxas_finais, color=['darkblue', 'blue', 'lightblue'])

plt.xticks(rotation=0)
plt.grid(axis='y', linestyle=':')
plt.yticks(range(0, 101, 10))
plt.title('Taxa de Sobrevivência por Porto de Embarque')
plt.ylabel('Sobreviventes (%)')

plt.show()
```
